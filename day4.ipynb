```javascript
{
  const stringToNumbersArray = (str) =>
    str.split(" ").map(Number).filter(Boolean);

  const cards = input.map((line) => {
    const values = line.split(": ")[1];
    const [winningNumbersString, myNumbersString] = values.split("|");
    const winningNumbers = stringToNumbersArray(winningNumbersString);
    const myNumbers = stringToNumbersArray(myNumbersString);
    console.log(myNumbers);
    let points = 0;
    for (const num of myNumbers) {
      if (!winningNumbers.includes(num)) continue;
      console.log(`match: ${num} in ${line.split(":")[0]}`);
      if (!points) {
        points = 1;
      } else {
        points = points * 2;
      }
    }
    console.log(`Points: ${points}\n`);

    return points;
  });

  cards.reduce((acc, points) => acc + points, 0);
}
```

## Second puzzle


```javascript
{
  const stringToNumbersArray = (str) =>
    str.split(" ").map(Number).filter(Boolean);

  const cards = input.reduce((acc, line) => {
    const values = line.split(": ")[1];
    const number = Number(line.split(":")[0].replace("Card ", ""));
    const [winningNumbersString, myNumbersString] = values.split("|");
    const winningNumbers = stringToNumbersArray(winningNumbersString);
    const myNumbers = stringToNumbersArray(myNumbersString);
    const matchingNumbers = myNumbers.filter((num) =>
      winningNumbers.includes(num)
    );

    acc[number] = { matchingNumbers, instances: 1 };
    return acc;
  }, {});

  // loop through the cards to see my winnings
  for (const number of Object.keys(cards)) {
    for (let i = 0; i < cards[number].instances; i++) {
      let index = Number(number);
      for (const x of cards[number].matchingNumbers) {
        index++;
        cards[index].instances++;
        // console.log(`Card ${number} wins us card ${index}`);
        // console.log(`    Card ${index} instances: ${cards[index].instances}`);
      }
    }
  }

  Object.values(cards).reduce((acc, { instances }) => acc + instances, 0);
}
```
